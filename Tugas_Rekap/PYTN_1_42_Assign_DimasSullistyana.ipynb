### Histogram

Histogram menunjukan data tingkat kejahatan di london dari tahun 2008 s/d 2016. 

```python
# df.loc[df["value"] > 10,["year","value"]]
df["year"].astype("str")
x = df.loc[df["value"] > 10,["year","value"]].groupby("year")["value"].sum()
x.plot(kind='hist', figsize=(8, 5))
plt.title('Histogram of London criminals') # add a title to the histogram
plt.ylabel('Summary Of Criminals') # add y-label
plt.xlabel('Years') # add x-label

# plt.show()
```
